#Data Visualization using Python, Matplotlib and Seaborn

Data visualization is the graphic representation of data. It involves producing images that communicate relationships among the represented data to viewers. Visualizing data is an essential part of data analysis and machine learning. We'll use Python libraries Matplotlib and Seaborn to learn and apply some popular data visualization techniques.

To do:

- Creating and customizing line charts using Matplotlib
- Visualizing relationships between two or more variables using scatter plots
- Studying distributions of variables using histograms & bar charts to 
- Visualizing two-dimensional data using heatmaps
- Displaying images using Matplotlib's `plt.imshow`
- Plotting multiple Matplotlib and Seaborn charts in a grid

```python
#Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns #It is build on top of matplotlib
%matplotlib inline
```

##Line Chart

The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines.

```python
#Create a sample data - Yield over the past 6 years
yields = [0.89, 0.9, 0.95, 0.98, 0.989, 0.991]
```

```python
plt.plot(yields);
```

```python
#Customize x - axis
years = range(2010, 2016)
years
```

```python
plt.plot(years, yields);
```

```python
#Add Labels
plt.plot(years, yields);
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare');
```

```python
#Plot Multiple Lines
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
#Add title and legends
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title("Crop Yields")
plt.legend(['Apples', 'Oranges']);
```

```python
#Line Markers
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')
plt.title('Crop Yield')
plt.legend(['Apples', 'Oranges']);
```

### Styling Lines and Markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line 
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot

```python
 #Styling Lines and Markers
plt.plot(years, apples, marker='s', c='darkolivegreen', ls='--')
plt.plot(years, oranges, marker='o', c='darkviolet', ls='--')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')
plt.title('Crop Yield')
plt.legend(['Apples', 'Oranges']);
```

```python
#Another way of passing arguments - fmt = '[marker][line][color]'
 #Styling Lines and Markers
plt.plot(years, apples, 's--r')
plt.plot(years, oranges, 'o--b')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')
plt.title('Crop Yield')
plt.legend(['Apples', 'Oranges']);
```

```python
#No line and only markers
plt.plot(years, apples, 'sr')
plt.plot(years, oranges, 'ob')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')
plt.title('Crop Yield')
plt.legend(['Apples', 'Oranges']);
```

```python
#Change figure size
plt.figure(figsize=(10,9))
plt.plot(years, apples, marker='s', c='darkred', ls='--')
plt.plot(years, oranges, marker='o', c='seagreen', ls='--')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare')
plt.title('Crop Yield')
plt.legend(['Apples', 'Oranges']);
```

##Scatter Plot

In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid.

```python
#Load Dataset
iris_df = sns.load_dataset('iris')
```

```python
iris_df
```

```python
iris_df.species.unique()
```

```python
#Visualize the relation between sepal length and sepal width.
sns.scatterplot(x=iris_df.sepal_length, y=iris_df.sepal_width);
```

```python
#Adding Hues
sns.scatterplot(
    x = iris_df.sepal_length,
    y = iris_df.sepal_width,
    hue = iris_df.species
);
```

```python
#Customize Seaborn figure

plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')
sns.scatterplot(
    x = iris_df.sepal_length,
    y = iris_df.sepal_width,
    hue = iris_df.species
);
```

```python
#Passing in pandas dataframe
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')
sns.scatterplot(
    x = 'sepal_length',
    y = 'sepal_width',
    hue = 'species',
    data = iris_df
);
```

## Histogram

A histogram represents the distribution of a variable by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin.

```python
#Plot sepal_width
plt.title('Distribution of Sepal Width')
plt.hist(iris_df.sepal_width);
```

```python
#Specify the number of bins
plt.title('Distribution of Sepal Width')
plt.hist(iris_df.sepal_width, bins = 5);
```

```python
#Specify the size of bins
import numpy as np
plt.title('Distribution of Sepal Width')
plt.hist(iris_df.sepal_width, bins = np.arange(2,5,0.25));
```

```python
setosa_df = iris_df[iris_df['species'] == 'setosa']
versicolor_df = iris_df[iris_df['species'] == 'versicolor']
virginica_df = iris_df[iris_df['species'] == 'virginica']
```

```python
#Multiple Histograms
plt.hist(setosa_df.sepal_width, alpha=0.5, bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=0.5, bins=np.arange(2,5,0.25))
plt.hist(virginica_df.sepal_width, alpha=0.5, bins = np.arange(2,5,0.25))
plt.title('Distribution of Sepal Width')
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
#Stacking multiple histograms on top of each other
plt.hist(
    [setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
    bins=np.arange(2,5,0.25),
    stacked=True
)
plt.title('Distribution of Sepal Width')
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

##Bar Chart

Bar charts are quite similar to line charts, i.e., they show a sequence of values. However, a bar is shown for each value, rather than points connected by lines. We can use the plt.bar function to draw a bar chart.

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
#Plotting a line chart
plt.plot(years,apples);
```

```python
#Plotting a bar chart for the same data
plt.bar(years,apples);
```

```python
#Plotting bar charts and line charts on the same graph
plt.plot(years,apples, c='darkred', marker='o', ls='--')
plt.bar(years,apples);
plt.title('Yield of Apples');
```

```python
#Stacking bars on top of one another

plt.bar(years, apples);
plt.bar(years, oranges, bottom=apples);
plt.title('Crop Yield');
plt.legend(['Apples', 'Oranges']);
```

##Bar Plots with Average

```python
#Load dataset
tips_df = sns.load_dataset('tips');
tips_df
```

```python
#Average total_bill by the week day
bill_average_df = tips_df.groupby('day')[['total_bill']].mean()
bill_average_df
```

```python
plt.bar(bill_average_df.index, bill_average_df.total_bill);
```

The lines cutting each bar represent the amount of variation in the values. For instance.

```python
#Doing the above steps directly using seaborn 
sns.barplot(x='day', y='total_bill', data=tips_df);
```

```python
#Specify hue to compare bar plots
sns.barplot(
    x = 'day',
    y =  'total_bill',
    hue='sex', 
    data = tips_df);
```

```python
#Making the bars horizontal by switching the axis
sns.barplot(
    y = 'day',
    x =  'total_bill',
    hue='sex', 
    data = tips_df);
```

##Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors.

```python
#Load dataset
flights_df = sns.load_dataset('flights')
flights_df
```

```python
#Draw a line plot
plt.plot(flights_df.passengers);
```

```python
#Represent data as matrix
flights_df = flights_df.pivot("month", "year", "passengers")
flights_df
```

```python
#Visualise Heatmap
sns.heatmap(flights_df);
plt.title("No. of Passengers (in thousands)");
```

```python
#Show value in each block
plt.figure(figsize=(12,8))
sns.heatmap(flights_df, annot=True, fmt='d', cmap='Blues');
plt.title("No. of Passengers (in thousands)");
```

##Plotting Multip Charts in a Grid

```python
#Plotting multiple subplots
fig, axes = plt.subplots(2,3, figsize=(16,8))
plt.tight_layout(pad=3);

#axes[0,0]
axes[0,0].plot(years, apples, marker='s', c='darkolivegreen', ls='--')
axes[0,0].plot(years, oranges, marker='o', c='darkviolet', ls='--')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare')
axes[0,0].set_title('Crop Yield')
axes[0,0].legend(['Apples', 'Oranges']);

#axes[0,1]
axes[0,1].set_title('Sepal Dimensions')
sns.scatterplot(
    x = 'sepal_length',
    y = 'sepal_width',
    hue = 'species',
    data = iris_df,
    ax = axes[0,1]
);

#axes[0,2]
axes[0,2].hist(
    [setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
    bins=np.arange(2,5,0.25),
    stacked=True
)
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);


#axes[1,0]
axes[1,0].plot(years,apples, c='darkred', marker='o', ls='--')
axes[1,0].bar(years,apples);
axes[1,0].set_title('Yield of Apples');


#axes[1,1]
axes[1,1].set_title('Average Bill')
sns.barplot(
    y = 'day',
    x =  'total_bill',
    hue='sex', 
    data = tips_df,
    ax=axes[1,1]);


#axes[1,2]
axes[1,2].set_title("No. of Passengers (in thousands)");
sns.heatmap(flights_df, cmap='Blues',ax=axes[1,2]);
```
